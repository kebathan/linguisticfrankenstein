```python
import pandas as pd
import numpy as np
```

```python
cognet_df = pd.read_csv("CogNet-v2.0.tsv", sep="\t", on_bad_lines="skip").dropna()
```

```python
cognet_df
```

```python
dravidian = ["tam", "kan", "tel", "mal"]
```

```python
mask1 = cognet_df['lang 1'].isin(["tam"])
mask2 = cognet_df['lang 2'].isin(["tam"])
tamil = pd.concat([cognet_df[mask1], cognet_df[mask2]], ignore_index=True, axis=0)

tamil
```

```python
mask1 = cognet_df['lang 1'].isin(["kan"])
mask2 = cognet_df['lang 2'].isin(["kan"])
kannada = pd.concat([cognet_df[mask1], cognet_df[mask2]], ignore_index=True, axis=0)

kannada
```

```python
mask1 = cognet_df['lang 1'].isin(["tel"])
mask2 = cognet_df['lang 2'].isin(["tel"])
telugu = pd.concat([cognet_df[mask1], cognet_df[mask2]], ignore_index=True, axis=0)

telugu
```

```python
mask1 = cognet_df['lang 1'].isin(["mal"])
mask2 = cognet_df['lang 2'].isin(["mal"])
malayalam = pd.concat([cognet_df[mask1], cognet_df[mask2]], ignore_index=True, axis=0)

malayalam
```

## Extracting Dravidian Languages

```python
mask1 = tamil['lang 1'].isin(dravidian)
tam_drav = tamil[mask1]
mask2 = tam_drav['lang 2'].isin(dravidian)
tam_drav = tam_drav[mask2]

tam_drav
```

```python
mask1 = telugu['lang 1'].isin(dravidian)
tel_drav = telugu[mask1]
mask2 = tel_drav['lang 2'].isin(dravidian)
tel_drav = tel_drav[mask2]

tel_drav
```

```python
mask1 = kannada['lang 1'].isin(dravidian)
kan_drav = kannada[mask1]
mask2 = kan_drav['lang 2'].isin(dravidian)
kan_drav = kan_drav[mask2]

kan_drav
```

```python
mask1 = malayalam['lang 1'].isin(dravidian)
mal_drav = malayalam[mask1]
mask2 = mal_drav['lang 2'].isin(dravidian)
mal_drav = mal_drav[mask2]

mal_drav
```

```python
drav_df = pd.concat([tam_drav, kan_drav, mal_drav, tel_drav], ignore_index=True, axis=0)
drav_df
```

```python
mask = drav_df["lang 1"] != drav_df["lang 2"]
drav_df = drav_df[mask].reset_index()
drav_df
```

```python
drav_df = drav_df.drop('concept id', axis=1)
drav_df = drav_df.drop('index', axis=1)

drav_df
```

```python
translit1 = drav_df.pop('translit 1')
drav_df.insert(2, 'translit 1', translit1)

drav_df
```

```python
test = drav_df.head(5)
test
```

```python
ord = {"tam":0, "kan":1, "mal":2, "tel":3}
drav_clean_test = np.array([["peyar", None, "per", None]])
```

```python
for index, row in test.iterrows():
    lang1 = row['lang 1']
    translit1 = row["translit 1"]

    if translit1 not in drav_clean_test[:, ord[lang1]]:
        new_word = np.array([None, None, None, None])
        new_word[ord[lang1]] = translit1
        drav_clean_test = np.vstack((drav_clean_test, new_word))
        idx = len(drav_clean_test) - 1    
    else:
        idx = list(drav_clean_test[:, ord[lang1]]).index(translit1)

    lang2 = row['lang 2']
    translit2 = row["translit 2"]

    if drav_clean_test[idx, ord[lang2]] == None:
        drav_clean_test[idx, ord[lang2]] = translit2
```

```python
ord = {"tam":0, "kan":1, "mal":2, "tel":3}
drav_clean = np.array([["peyar", None, "per", None]])
drav_clean
```

```python
for index, row in drav_df.iterrows():
    lang1 = row['lang 1']
    translit1 = row["translit 1"]

    if translit1 not in drav_clean[:, ord[lang1]]:
        new_word = np.array([None, None, None, None])
        new_word[ord[lang1]] = translit1
        drav_clean = np.vstack((drav_clean, new_word))
        idx = len(drav_clean) - 1    
    else:
        idx = list(drav_clean[:, ord[lang1]]).index(translit1)

    lang2 = row['lang 2']
    translit2 = row["translit 2"]

    if drav_clean[idx, ord[lang2]] == None:
        drav_clean[idx, ord[lang2]] = translit2
```

```python
drav_clean
```

```python
drav_final = pd.DataFrame({'tam': drav_clean[:, 0], 'kan': drav_clean[:, 1], 'mal': drav_clean[:, 2], 'tel': drav_clean[:, 3]})
drav_final = drav_final.fillna(value=np.nan)
drav_final
```

```python
drav_final.to_csv('drav_final.csv', index=False)
```
