```python
# import statements
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# import dravidian dataset
drav_final = pd.read_csv("drav_final.csv")
drav_final
```

```python
# remove all inputs with NaN in tamil column
tamil_data = drav_final[drav_final["tam"].notnull()]
tamil_data
```

```python
# connect multi-word inputs with _
tamil_data['kan'] = tamil_data['kan'].apply(lambda x: x.replace(" ", "_") if isinstance(x, str) else x)
tamil_data['tel'] = tamil_data['tel'].apply(lambda x: x.replace(" ", "_") if isinstance(x, str) else x)
tamil_data['mal'] = tamil_data['mal'].apply(lambda x: x.replace(" ", "_") if isinstance(x, str) else x)
tamil_data['tam'] = tamil_data['tam'].apply(lambda x: x.replace(" ", "_") if isinstance(x, str) else x)

tamil_data
```

```python
# replace null values with NaN
tamil_data = tamil_data.fillna("NaN")
tamil_data
```

```python
# separate into training, validation, and testing data
train, temp = train_test_split(tamil_data, test_size=0.2, random_state=42)
valid, test = train_test_split(temp, test_size=0.5, random_state=42)

train, valid, test
```

```python
# combine kannada, malayalam, and telugu into one word
def combine_inputs(row):
    combined = []
    for i in range(max(len(row["kan"]), len(row["mal"]), len(row["tel"]))):
        if i < len(row["kan"]):
            if row["kan"] != "NaN":
                combined.append(row["kan"][i])

        if i < len(row["mal"]):
            if row["mal"] != "NaN":
                combined.append(row["mal"][i])

        if i < len(row["tel"]):
            if row["tel"] != "NaN":
                combined.append(row["tel"][i])

    return "".join(combined)
```

```python
# training data
train_inputs = train.apply(combine_inputs, axis=1)
train_targets = train['tam']

# validation data
valid_inputs = valid.apply(combine_inputs, axis=1)
valid_targets = valid['tam']

# testing data
test_inputs = test.apply(combine_inputs, axis=1)
test_targets = test['tam']
```

```python
# save files
train_inputs.to_csv('..\\data\\train_inputs.txt', index=False, header=False)
train_targets.to_csv('..\\data\\train_targets.txt', index=False, header=False)

valid_inputs.to_csv('..\\data\\valid_inputs.txt', index=False, header=False)
valid_targets.to_csv('..\\data\\valid_targets.txt', index=False, header=False)

test_inputs.to_csv('..\\data\\test_inputs.txt', index=False, header=False)
test_targets.to_csv('..\\data\\test_targets.txt', index=False, header=False)
```

```python
# combine kannada, malayalam, and telugu into many spaces
def combine_inputs2(row):
    combined = []
    for i in range(max(len(row["kan"]), len(row["mal"]), len(row["tel"]))):
        if i < len(row["kan"]):
            if row["kan"] != "NaN":
                combined.append(row["kan"][i])

        if i < len(row["mal"]):
            if row["mal"] != "NaN":
                combined.append(row["mal"][i])

        if i < len(row["tel"]):
            if row["tel"] != "NaN":
                combined.append(row["tel"][i])

    return " ".join(combined)
```

```python
def split_tamil(row):
    return " ".join(list(row))
```

```python
# training data
train_inputs2 = train.apply(combine_inputs2, axis=1)
train_targets2 = train['tam'].apply(split_tamil)

# validation data
valid_inputs2 = valid.apply(combine_inputs2, axis=1)
valid_targets2 = valid['tam'].apply(split_tamil)

# testing data
test_inputs2 = test.apply(combine_inputs2, axis=1)
test_targets2 = test['tam'].apply(split_tamil)
```

```python
# save files
train_inputs2.to_csv('..\\data\\train_inputs2.txt', index=False, header=False)
train_targets2.to_csv('..\\data\\train_targets2.txt', index=False, header=False)

valid_inputs2.to_csv('..\\data\\valid_inputs2.txt', index=False, header=False)
valid_targets2.to_csv('..\\data\\valid_targets2.txt', index=False, header=False)

test_inputs2.to_csv('..\\data\\test_inputs2.txt', index=False, header=False)
test_targets2.to_csv('..\\data\\test_targets2.txt', index=False, header=False)
```

```python
combined_words = pd.concat([train_inputs, train_targets, valid_inputs, valid_targets, test_inputs, test_targets], ignore_index=True, axis=0)
combined_words
```

```python
import sentencepiece as spm

# Train SentencePiece model
spm.SentencePieceTrainer.train(input='..\\data\\trial2\\combined.txt',
                               model_prefix='..\\data\\trial2\\spm',
                               vocab_size=10000)
```

```python
# Load the trained SentencePiece model
sp = spm.SentencePieceProcessor(model_file="..\\data\\spm.model")

# Tokenize the training dataset (inputs and targets)
with open("..\\data\\train_inputs.txt", "r", encoding="utf-8") as infile, open("..\\data\\train_inputs_spm.txt", "w", encoding="utf-8") as outfile:
    for line in infile:
        tokens = sp.encode(line.strip(), out_type=str)
        outfile.write(" ".join(tokens) + "\n")

with open("..\\data\\train_targets.txt", "r", encoding="utf-8") as infile, open("..\\data\\train_targets_spm.txt", "w", encoding="utf-8") as outfile:
    for line in infile:
        tokens = sp.encode(line.strip(), out_type=str)
        outfile.write(" ".join(tokens) + "\n")

# Tokenize the validation dataset (inputs and targets)
with open("..\\data\\valid_inputs.txt", "r", encoding="utf-8") as infile, open("..\\data\\valid_inputs_spm.txt", "w", encoding="utf-8") as outfile:
    for line in infile:
        tokens = sp.encode(line.strip(), out_type=str)
        outfile.write(" ".join(tokens) + "\n")

with open("..\\data\\valid_targets.txt", "r", encoding="utf-8") as infile, open("..\\data\\valid_targets_spm.txt", "w", encoding="utf-8") as outfile:
    for line in infile:
        tokens = sp.encode(line.strip(), out_type=str)
        outfile.write(" ".join(tokens) + "\n")

# Tokenize the testing dataset (inputs only, since targets are not needed for translation)
with open("..\\data\\test_inputs.txt", "r", encoding="utf-8") as infile, open("..\\data\\test_inputs_spm.txt", "w", encoding="utf-8") as outfile:
    for line in infile:
        tokens = sp.encode(line.strip(), out_type=str)
        outfile.write(" ".join(tokens) + "\n")
```

```python
# Re-encode the input file to UTF-8
input_file = "..\\data\\single_test_input.txt"
output_file = "..\\data\\single_test_input_utf8.txt"

with open(input_file, "rb") as infile, open(output_file, "w", encoding="utf-8") as outfile:
    for line in infile:
        decoded_line = line.decode("utf-8", errors="replace")
        outfile.write(decoded_line)
```
